##### First Bokeh Plot

```python
from bokeh.plotting import figure
from bokeh.io import output_file, show
```

```python
#Create some datasets
x=[0,1,2,3,4]
y=[5,6,7,8,9]

#Prepare the output file
output_file("Line.html")

#Also create a figure instance
f=figure()

#Create a line plot
f.line(x,y)

#Now show the plot
show(f)
```

```python
#We can also the glyphs available by using 
dir(f)
```

```python
#Lets change the graph and use a triangle as the marker
f.triangle(x=x,y=y)
show(f)
```

```python
#We can also do a circle
f.circle(size=10,x=x,y=y)
show(f)
```

```python
#The code we run sequentially are used to overlay the commands and 
#hence should be used cautiously

f1=figure()
f1.triangle(x,y)
show(f1)

#This will only show triangle points on the graphs
```
