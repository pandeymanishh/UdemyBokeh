### Custom HTML Tooltips with Images

```python
#Get the libraries
import numpy as np
import pandas as pd
from pydataset import data
#Import the figure function
from bokeh.plotting import figure
#Import output utilities
from bokeh.io import output_file, show
#Import Plot Utilities
from bokeh.models import Range1d,PanTool,ResetTool,CrosshairTool,HoverTool,ColumnDataSource,LassoSelectTool
#Import COlor Palettes
from bokeh.palettes import d3,mpl,brewer
```

```python
#Get the iris dataset
flowers=data('iris')
flowers.columns=[i.lower().replace('.','') for i in flowers.columns]
```

```python
#Now create the plot

#First add the output object
output_file("IrisCustom.html")

#Now add the figure object
f=figure(tools=[PanTool(),ResetTool(),CrosshairTool(),LassoSelectTool()]  #Select the tools you want to see
         ,toolbar_location="below",toolbar_sticky=False)


#You can also remove the logo
f.toolbar.logo=None

#No minor grid lines
f.grid.minor_grid_line_color=None
#Add Axis labels
f.xaxis.axis_label="Petal Length"
f.yaxis.axis_label="Petal Width"

#Make it italices
f.axis.axis_label_text_font_style="italic"

#Increase the font size
f.axis.axis_label_text_font_size="20px"

#Get the 


#Create the color palette
vals=['setosa','versicolor','virginica']
clr=d3['Category10'][3]

imgsrc={'setosa':'F:\\Python Stuff\\Python Learning\\UdemyBokeh\\Img\\setosa.png'
        ,'versicolor':'F:\\Python Stuff\\Python Learning\\UdemyBokeh\\Img\\versicolor.png'
        ,'virginica':'F:\\Python Stuff\\Python Learning\\UdemyBokeh\\Img\\virginica.png'}

flowers['img']=[imgsrc[j] for j in flowers.species]


for i,j in zip(vals,clr):
    
    #Create and filter data
    dt=flowers.loc[flowers.species==i]
    dt1=ColumnDataSource(dt)

    #Get the basic plot done
    f.circle(x="petallength"
             ,y="petalwidth"
             ,size="sepallength"
            ,fill_alpha=0.5   #Dim the fill color
             ,color=j
             ,legend=i
             ,line_dash='solid'
             ,muted_color=j    #Add functionality for legend selection
             ,muted_alpha=0.1
             ,source=dt1)

#Add Custom hover tool
#hover = HoverTool(tooltips=[("Element ","@name"),("Metal ","@metal"),("State ","@standardstate")])

#f.add_tools(hover)

##Adding custom html for hovertools
hover=HoverTool(tooltips="""
                    <div>
                      <div>
                        <img 
                            src="file://@img" height="42" alt="@species" width="42"
                            style="float:left; margin:0px 15px 15px 0px;"
                            border="2">
                        </img>
                      </div>
                      <div>
                        <span style="font-size:20px; font-weight: bold;text-transform:capitalize;">@species
                        </span><br>
                      </div>
                      <div>
                        <span style="font-size:10px; color: red;">Petal Length: @petallength</span><br>
                        <span style="font-size:10px; color: green;">Petal Width: @petalwidth</span>       
                      </div>
                    </div>
                    """)

f.add_tools(hover)


#Configure the legend poliy
f.legend.location="top_left"
f.legend.click_policy="mute"
f.legend.border_line_color=None


show(f)

```
